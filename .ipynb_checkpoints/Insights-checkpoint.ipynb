#### lets replace the null values by group cities together and putting values from the other rows as taking mode of the values

```python
def fill_with_mode(group):
    mode_value = group['Cuisines'].mode().iloc[0] if not group['Cuisines'].mode().empty else np.nan
    group['Cuisines'].fillna(mode_value, inplace=True)
    return group

# Apply the function to each group
df1 = df1.groupby('City').apply(fill_with_mode)

# Reset index to restore the original DataFrame structure
df1.reset_index(drop=True, inplace=True)
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
#Now there is still one Null Value which we will replace with the rows having similar country code
df1 = df1.groupby('Country Code').apply(fill_with_mode)
```

```python
df1.isnull().sum()
```

```python
#Now our Dataset is cleaned up of Null Values 
#Now lets read another csv file having country code and country Name
df2=pd.read_excel('Country-Code.xlsx')
df2.head()
```

```python
#So In this dataset we have two columns column 1 shows Country Code and Column 2 Shows Country so we can join this with table df1 with help of join key Country Code to get the associated Country
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
```

```python
Final_df = pd.merge(df1, df2, on='Country Code', how='left') 
#Here we are doing left only because we want all the rows from the df1 table weather we have matching country code in df2 or not
```

```python
Final_df.head()
```

```python
Final_df.drop('Country Code', axis=1, inplace=True)
```

```python
rating = Final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
rating
```

## Few Observations:
1. When the rating is between 4.5 AND 4.9 it is Excellent
